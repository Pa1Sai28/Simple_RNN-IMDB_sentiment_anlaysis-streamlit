```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'i will get a job this first quarter']
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
sent
```

```python
## one hot representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
## the pad_sequences and padding helps in representing a zero as a balance to the number of words
sent_length=8
embdedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embdedded_docs)
```

```python
## feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```python
model.predict(embdedded_docs)
```
